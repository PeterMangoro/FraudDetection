````python
# import pandas
import pandas as pd

# load dataset

df = pd.read_csv(file_path)

# Quick look at the dataset
print("Shape:", df.shape)   # rows, columns
print("\nColumns:", df.columns.tolist())

# View first few rows
print("\nSample rows:")
print(df.head())

# Dataset info
print("\nInfo:")
print(df.info())

# Summary stats (numerical only)
print("\nSummary statistics:")
print(df.describe())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
````

## 🧹 Data Preprocessing & Feature Engineering

Now that we understand our data, let's clean and prepare it for modeling. This involves several important steps:

### 🔧 Preprocessing Steps:
1. **Data Cleaning**: Remove unnecessary columns and clean column names
2. **DateTime Processing**: Extract temporal features from transaction timestamps
3. **Feature Creation**: Build meaningful features for fraud detection
4. **Encoding**: Convert categorical variables to numerical format
5. **Scaling**: Normalize numerical features for machine learning

### ⚠️ Important Notes:
- The original `time_of_day` column contained all NaN values, so we'll create our own time-of-day feature
- We'll engineer features that capture user behavior patterns and risk indicators


````python
#  delete unnamed column

df.head()

#  clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# drop time_of_day(morning,_afternoon,_evening,_night)
df = df.drop(['time_of_day(morning,_afternoon,_evening,_night)'], axis=1)
df = df.drop(['unnamed:_0'], axis=1)

# convert datetime to pandas datetime
df['datetime'] = pd.to_datetime(df['datetime'])

#  extract useful features
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['day_of_week'] = df['datetime'].dt.day_name()

# set time of day (morning,afternoon, evening, night) include 0 bin

def map_time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

# Create new column
df["time_of_day"] = df["hour"].apply(map_time_of_day)

# check missing values
print('missing values', df.isnull().sum())



df.head()
````

### 🏗️ Advanced Feature Engineering

Here we create sophisticated features that capture behavioral patterns and risk indicators commonly associated with fraudulent transactions:

#### 📈 **User Behavior Features:**
- **Daily Transaction Count**: Number of transactions per user per day (high frequency may indicate fraud)
- **Average Transaction Amount**: User's typical transaction size (deviations may be suspicious)
- **Amount Deviation**: Difference between current transaction and user's average (large deviations are risky)

#### 🚨 **Risk Indicators:**
- **Night Transaction Flag**: Transactions during night hours (often associated with fraud)
- **SIM/Multiple Account Risk Score**: Combined risk from SIM swapping and multiple accounts
- **Foreign High Amount**: High-value transactions to foreign numbers (high-risk combination)
- **Location Diversity**: Number of unique locations per user (unusual location patterns)

#### 💡 **Why These Features Matter:**
- **Behavioral Anomalies**: Fraudsters often exhibit different patterns than legitimate users
- **Risk Amplification**: Combining multiple risk factors increases detection accuracy
- **Temporal Patterns**: Fraud often occurs at unusual times


````python
# Preparing for Isolation Forest
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['transaction_type', 'location', 'device_type', 'network_provider', 'user_type', 'time_of_day', 'day_of_week']

# Initialize encorder
encoder = OneHotEncoder(sparse_output=False, drop='first')  # drop first to avoid multicollinearity


# Fit and transform
encoded_data = encoder.fit_transform(df[categorical_cols])

# Create a new DataFrame with the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

#  Merge with original dataframe (drop original columns)
df_model = pd.concat([df.drop(columns=categorical_cols), encoded_df], axis=1)

df_model.head()
````

### 📊 Data Scaling & Preparation

Before training our machine learning model, we need to scale our numerical features to ensure all variables contribute equally to the model:

#### 🎯 **Why Scaling is Important:**
- **Feature Equality**: Prevents features with larger scales from dominating the model
- **Algorithm Requirements**: Isolation Forest performs better with normalized data
- **Consistent Interpretation**: Makes feature importance comparisons meaningful

#### 📈 **Features Being Scaled:**
- Transaction amounts and temporal features
- Risk indicators and behavioral metrics
- User-specific statistics and deviations


````python
## Feature Engineering

# Transactions per user
df_model['date'] = df_model['datetime'].dt.date
user_txn_count = df_model.groupby(['user_id', 'date'])['transaction_id'].count().reset_index()
user_txn_count.rename(columns={'transaction_id': 'txn_count_count_per_day'}, inplace=True)

# Drop the potentially existing column before merging
if 'txn_count_count_per_day' in df_model.columns:
    df_model = df_model.drop('txn_count_count_per_day', axis=1)

# Merge back to main dataframe
df_model = pd.merge(df_model, user_txn_count, on=['user_id', 'date'], how='left')

# Average transaction amount per user
user_avg_txn_amt = df_model.groupby(['user_id'])['amount'].mean().reset_index()
user_avg_txn_amt.rename(columns={'amount': 'avg_txn_amt'}, inplace=True)

# Drop the potentially existing column before merging
if 'avg_txn_amt' in df_model.columns:
    df_model = df_model.drop('avg_txn_amt', axis=1)

# Merge
df_model = pd.merge(df_model, user_avg_txn_amt, on=['user_id'], how='left')

# Amount deviation (current txn vs user avg)
df_model['txn_amt_deviation'] = df_model['amount'] - df_model['avg_txn_amt']

# Signal midnight txns (usualy suspicious)
df_model['is_night_txn'] = df['time_of_day'].apply(lambda x: 1 if x == 'night' else 0)

#  Risk Score for swapped SIM and multiple accounts
df_model['sim_multiple_risk_score'] = df_model['is_sim_recently_swapped'] + df_model['has_multiple_accounts']

# Foreign number high amount
df_model['foreign_high_amt'] = df_model['is_foreign_number'] * df_model['amount']

# Count number of locations per user
user_location_count = df.groupby(['user_id'])['location'].nunique().reset_index()
user_location_count.rename(columns={'location': 'unique_location_count'}, inplace=True)

# Drop the potentially existing column before merging
if 'unique_location_count' in df_model.columns:
    df_model = df_model.drop('unique_location_count', axis=1)

# Merge
df_model = pd.merge(df_model, user_location_count, on=['user_id'], how='left')
````

## 🤖 Machine Learning Model: Isolation Forest

Now we'll implement our fraud detection model using the **Isolation Forest** algorithm, which is particularly well-suited for anomaly detection in unsupervised scenarios.

### 🌲 **Why Isolation Forest?**

#### ✅ **Advantages:**
- **Unsupervised**: Works without labeled fraud data (common in real-world scenarios)
- **Efficient**: Fast training and prediction on large datasets
- **Interpretable**: Provides anomaly scores that can be interpreted as risk levels
- **Robust**: Less sensitive to outliers compared to other algorithms
- **Scalable**: Handles high-dimensional data well

#### 🔧 **Model Parameters:**
- **n_estimators=200**: Number of isolation trees (more trees = more stable predictions)
- **contamination=0.02**: Expected proportion of anomalies (2% - reasonable for fraud detection)
- **max_samples='auto'**: Number of samples to draw from dataset for each tree
- **random_state=42**: Ensures reproducible results

### 📊 **How It Works:**
1. **Isolation**: Creates random isolation trees that separate normal data points
2. **Anomaly Scoring**: Anomalies require fewer splits to isolate (shorter paths in trees)
3. **Decision Function**: Returns scores where lower values indicate higher anomaly probability
4. **Classification**: Maps scores to binary fraud predictions (1 = fraud, 0 = normal)


````python
# Scaling Numerical Features
from sklearn.preprocessing import StandardScaler

numerical_cols = [
    'amount','month','day','hour',
    'is_foreign_number', 'is_sim_recently_swapped', 'has_multiple_accounts',
    'txn_count_count_per_day','avg_txn_amt','txn_amt_deviation',
    'is_night_txn','sim_multiple_risk_score','foreign_high_amt',
    'unique_location_count'
    ]


# Initialize scaler
scaler = StandardScaler()
df_model[numerical_cols] = scaler.fit_transform(df_model[numerical_cols])

print(df_model.shape)
df_model.head()
````

## 📈 Model Results & Performance Analysis

Let's analyze our model's performance and understand what patterns it has identified as potentially fraudulent.

### 📊 **Detection Summary:**
Our Isolation Forest model has identified **200 transactions (2%)** as potential fraud cases. This is within the expected range for fraud detection systems.

#### 🎯 **Key Metrics:**
- **Total Transactions**: 10,000
- **Predicted Fraud**: 200 (2.0%)
- **Predicted Normal**: 9,800 (98.0%)
- **Contamination Rate**: Matches our model parameter (2%)

### 🔍 **What This Means:**
- The model has flagged transactions that exhibit unusual patterns compared to the majority
- These anomalies could represent various types of fraudulent activity
- The 2% rate is reasonable for fraud detection (industry standard is typically 1-5%)


````python
from sklearn.ensemble import IsolationForest

# Select model-ready features exclude IDs and datetime columns
features_model = [col for col in df_model.columns if col not in ['transaction_id', 'user_id', 'datetime', 'date']]

X = df_model[features_model]

# Initialize model
iso = IsolationForest(
    n_estimators=200,
    max_samples='auto',
    contamination=0.02,
    random_state=42)

# Fit model
iso.fit(X)

# Predict anomalies
df_model['anomaly_score'] = iso.decision_function(X)
df_model['anomaly_label'] = iso.predict(X)

df_model['is_fraud_predicted'] = df_model['anomaly_label'].map({1:0,-1:1})

# Merge results back to original df
df['anomaly_score'] = df_model['anomaly_score']
# df['anomaly_label'] = df_model['anomaly_label']
df['is_fraud_predicted'] = df_model['is_fraud_predicted']

print(df['is_fraud_predicted'].value_counts())

# View top suspicious transactions
df[['transaction_id','user_id','amount','anomaly_score','is_fraud_predicted']].sort_values('anomaly_score').head(10)
````
